```python
#Os preços aumentam conforme o tamanho da casa.
```

```python
sns.scatterplot(data=df, x="tamanho_m2", y="preco")
plt.title("Tamanho da casa vs Preço")
plt.show()
```

```python
X = df[["tamanho_m2"]]
y = df["preco"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

pipeline.fit(X, y)
```

```python
import pandas as pd

nova_casa = pd.DataFrame(
    [[85]],
    columns=["tamanho_m2"]
)

preco_previsto = pipeline.predict(nova_casa)
preco_previsto
```
